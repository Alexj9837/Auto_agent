```python
from Moutain.model import Mountain
import matplotlib.pyplot as plt
import pandas as pd


n_robots = 3
n_Patient = 3
n_healers = 3
seed = 123
width = 50
height = 50
n_steps = 300

m = Mountain(n_robots, n_Patient, n_healers,
             width=width, height=height, seed=seed)

for i in range(n_steps):
    m.step()

m_df = m.datacollector.get_model_vars_dataframe()
ag_df = m.datacollector.get_agent_vars_dataframe()
```

# Battery drain for each agent

A brief elplaination of the below plots, the first three track the battery of the searching robot and the next three are of the healing drone.

```python
ag_df = ag_df.dropna()
ag_df.groupby("AgentID").plot(y="battery")
```
